## Figure 1

```python
nbMariages18 = len(data18)
print("nb mariage en 2018 = " + str(nbMariages18))
nbMariages19 = len(data19)
print("nb mariage en 2019 = " + str(nbMariages19))
```

```python
# Nombre de mariages par mois sur les années 2018 et 2019
nb_mariage_month18 = []
nb_mariage_month19 = []

nb_mariage_month18_sameSex = []
nb_mariage_month19_sameSex = []

for month_index in range(12):
  nb_mariage_month18.append(data18[(data18.MMAR == month_index+1) & (data18.SEXE1 != data18.SEXE2)].count()[0])
  nb_mariage_month19.append(data19[(data19.MMAR == month_index+1) & (data19.SEXE1 != data19.SEXE2)].count()[0])

  nb_mariage_month18_sameSex.append(data18[(data18.MMAR == month_index+1) & (data18.SEXE1 == data18.SEXE2)].count()[0])
  nb_mariage_month19_sameSex.append(data19[(data19.MMAR == month_index+1) & (data19.SEXE1 == data19.SEXE2)].count()[0])

print(nb_mariage_month18)
print(nb_mariage_month18_sameSex)
print()
print(nb_mariage_month19)
print(nb_mariage_month19_sameSex)
```

```python
months = ["Janvier", "Février", "Mars", "Avril",
          "Mai", "Juin", "Juillet", "Aout", "Septembre", "Octobre", "Novembre", "Décembre"]

fig,ax = plt.subplots()

ax.plot(list(range(1,13)), 
        nb_mariage_month18, 
        color="orange",
        marker="o", markerfacecolor="orange", markeredgecolor="white",
        label="2018 (sexe différent)")

ax.plot(list(range(1,13)), 
        nb_mariage_month19, 
        color="green",
        marker="o", markerfacecolor="green", markeredgecolor="white",
        label="2019 (sexe différent)")

ax.plot(list(range(1,13)), 
        nb_mariage_month18_sameSex, 
        color="orange", linestyle=(0, (5,5)),
        marker="o", markerfacecolor="orange", markeredgecolor="white",
        label="2018 (même sexe)")

ax.plot(list(range(1,13)), 
        nb_mariage_month19_sameSex, 
        color="green", linestyle=(5, (5,5)),
        marker="o", markerfacecolor="green", markeredgecolor="white",
        label="2019 (même sexe)")

ax.set_xticks(list(range(1,13)))
ax.set_xticklabels(months)

# Rotate the tick labels and set their alignment.
plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")

ax.set_title("Nombre de mariages par mois sur les années 2018 et 2019", fontsize=16)
ax.legend(loc='upper left', frameon=True, fontsize=14)
ax.grid(1)

for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(14);

fig.set_size_inches(14, 9)

plt.savefig("images/fig1_nb_mariage_mois.pdf")
plt.show();
```

## Figures 2 et 3

```python
# Conversion du numéro de département en int plutôt que string
data18['DEPMAR'] = pd.to_numeric(data18['DEPMAR'],errors='coerce')
data18 = data18.replace(np.nan, 0, regex=True)
data18['DEPMAR'] = data18['DEPMAR'].astype(int)
```

### 2018

```python
months = ["Janvier", "Février", "Mars", "Avril",
              "Mai", "Juin", "Juillet", "Aout", "Septembre", "Octobre", "Novembre", "Décembre"]
days = ["Lundi", "Mardi", "Mercredi",
           "Jeudi", "Vendredi", "Samedi", "Dimanche"]
                    
nb_mariage_2018 = []
for nb_day in range(len(days)):
  nb_mariage_days = []
  for nb_month in range(len(months)):
    nb_mariage_days.append(data18[(data18.JSEMAINE == nb_day+1) & (data18.MMAR == nb_month+1)].count()[0])
  nb_mariage_2018.append(nb_mariage_days)

fig, ax = plt.subplots()
im = ax.imshow(nb_mariage_2018, cmap="YlGn")


plt.colorbar(im,fraction=0.04, pad=0.04)

# We want to show all ticks...
ax.set_xticks(np.arange(len(months)))
ax.set_yticks(np.arange(len(days)))
# ... and label them with the respective list entries
ax.set_xticklabels(months)
ax.set_yticklabels(days)

# Rotate the tick labels and set their alignment.
plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")

# Loop over data dimensions and create text annotations.
for i in range(len(days)):
    for j in range(len(months)):
      if(i == 5):
        pourcentage = round(nb_mariage_2018[i][j]/data18[(data18.JSEMAINE == 6)].count()[0] * 100,2)
        text = ax.text(j, i, pourcentage,
                       ha="center", va="center", color="darkorange", fontweight='bold')

ax.set_title("Répartition des mariages de 2018 par jour et par mois (en % pour le samedi)")
fig.tight_layout()

fig.set_size_inches(10.5, 5.5)


plt.savefig("images/fig2_heatmap2018.pdf", dpi=100)
plt.show()
```

```python
# https://stackoverflow.com/a/22845857

n_year = 2
n_col = len(df18.columns) 
n_ind = len(df18.index)
#ax = plt.subplot(111)
fig, ax = plt.subplots()

ax = df18.plot(kind="bar",
               linewidth=0,
               stacked=True,
               ax=ax,
               legend=False,
               grid=False,
               colormap="Set3")  # make bar plot

ax = df19.plot(kind="bar",
               linewidth=0,
               stacked=True,
               ax=ax,
               legend=False,
               grid=False,
               colormap="Set3")

hatch_value = "/"
plt.rcParams['hatch.linewidth'] = 0.5
plt.rcParams['hatch.color'] = "white"

h,l = ax.get_legend_handles_labels() # get the handles we want to modify
for i in range(0, n_year * n_col, n_col): # len(h) = n_col * n_year
    for j, pa in enumerate(h[i:i+n_col]):
        for rect in pa.patches: # for each index
            rect.set_x(rect.get_x() + 1 / float(n_year + 1) * i / float(n_col))
            rect.set_hatch(hatch_value * int(i / n_col)) #edited part
            rect.set_width(1 / float(n_year + 1))

ax.set_xticks((np.arange(0, 2 * n_ind, 2) + 1 / float(n_year + 1)) / 2.)
ax.set_xticklabels(df18.index, rotation = 0)
title = "Répartition des mariés selon leur état matrimonial, par possession ou non d'un enfant en commun avant le mariage"
ax.set_title(title, fontsize=16)

# Add invisible data to add another legend
n=[]        
for i in range(n_year):
    n.append(ax.bar(0, 0, color="gray", hatch=hatch_value * i))

l1 = ax.legend(h[:n_col], l[:n_col], loc=[0.75, 0.7], fontsize=14, prop={'size': 12})
l2 = plt.legend(n, ["2018", "2019"], loc=[0.75, 0.6], fontsize=14, prop={'size': 12}) 
ax.add_artist(l1)

for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(14);

fig.set_size_inches(14, 9)
plt.savefig("images/fig5_repartition.pdf")
```
